```python
import pandas as pd
```

```python
dataset=pd.read_csv("LS_2.0.csv")
```

```python
dataset.info()
```

```python
dataset.rename(columns={'CRIMINAL\nCASES':'CRIMINAL CASES','GENERAL\nVOTES':'GENERAL VOTES','POSTAL\nVOTES':'POSTAL VOTES'},inplace=True)
```

```python
dataset
```

```python
dataset.EDUCATION.unique()
```

```python
dataset['EDUCATION'].replace(to_replace=['Illiterate','Not Available','Others'],value='Illiterate',inplace=True)
dataset['EDUCATION'].replace(to_replace=['12th Pass','10th Pass','8th Pass','5th Pass','Literate'],value='scholl level',inplace=True)
dataset['EDUCATION'].replace(to_replace=['Graduate Professional','Graduate'],value='Graduate',inplace=True)
dataset['EDUCATION'].replace(to_replace=['Post Graduate','Post Graduate\n','Doctorate'],value='Post Graduate/Doctorate',inplace=True)
```

```python
dataset.EDUCATION.unique()
```

```python
dataset
```

```python
dataset['CRIMINAL CASES']=dataset['CRIMINAL CASES'].replace({'Not available':0})
dataset['CRIMINAL CASES']=dataset['CRIMINAL CASES'].fillna(0)
```

```python
dataset.isnull().sum()
```

```python
sb.boxplot(x='EDUCATION',y='GENERAL VOTES',data=dataset)
plt.show()
```
